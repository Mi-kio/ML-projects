```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

# Example - 2 dollar signs

## $$ f(x) = x^2 + x + 1$$

```python
def f(x):
    return x*x + x + 1
```

```python
x_l = np.linspace(start = -3, stop = 3, num = 10)
x_l
```

```python
# plot
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_l,f(x_l))
plt.show()
```

```python
# we are making this plot to get minimum possible values

```

# slope and derivatives

```python
# plot
plt.figure(figsize=[15,5])

# 1st chart
plt.subplot(1,2,1)
plt.title('cost func')
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_l,f(x_l),color='blue',linewidth=3)

# 2nd chart

plt.subplot(1,2,2)
plt.title('slope of cost func')
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()
plt.xlim([-2,3])
plt.ylim([-3,6])
plt.plot(x_l,f(x_l),color='skyblue',linewidth=5)

plt.show()
```
