## $$ f(x) = x^2 + x + 1$$

```python
def f(x):
    return x*x + x + 1
```

```python
x_l = np.linspace(start = -3, stop = 3, num = 10)
x_l
```

```python
# plot
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_l,f(x_l))
plt.show()
```

```python
# we are making this plot to get minimum possible values

def df(x):
    return 2*x + 1
```

# slope and derivatives

```python
# plot
plt.figure(figsize=[15,5])

# 1st chart
plt.subplot(1,2,1)
plt.title('cost func')
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_l,f(x_l),color='blue',linewidth=3)

# 2nd chart

plt.subplot(1,2,2)
plt.title('slope of cost func')
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()
plt.xlim([-2,3])
plt.ylim([-3,6])
plt.plot(x_l,df(x_l),color='skyblue',linewidth=5)

plt.show()
```

## python loops and gradient descent

```python
# python for loop
for n in range (5):
    print('hello world')
```

```python
# gradient descent
new_x = 3
prev_x = 0
step_multi = 0.1

x_li  = [new_x]
slope_li = [df(new_x)]

preci = 0.0001


for n in range(500):
    prev_x = new_x
    gradient = df(prev_x)
    # stored slope in gradient
    # the further we are from minimum the steeper we are
    # steep 0 means small error, large value of steep means big error
    new_x = prev_x - step_multi * gradient
    
    step_size = abs(new_x - prev_x)
    print(step_size)
    
    x_li.append(new_x)
    slope_li.append(df(new_x))
    
    if step_size < preci:
        print('loop ran these many times ',n)
        break;

print('local minimum occurs at' , new_x)
print('slope of df(x) value at this point is :',df(new_x))
print('new value of new_x is: ',f(new_x))
```

```python
# plot
plt.figure(figsize=[20,5])

# 1st chart
plt.subplot(1,3,1)
plt.title('cost func')
plt.xlim([-3,3])
plt.ylim([0,8])
plt.xlabel('X',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_l,f(x_l),color='blue',linewidth=3,alpha=0.7)

values = np.array(x_li)
plt.scatter(x_li,f(values),color='red',s=100,alpha=0.6)

# 2nd chart

plt.subplot(1,3,2)
plt.title('slope of cost func')
plt.xlabel('X',fontsize=16)
plt.ylabel('df(x)',fontsize=16)
plt.grid()
plt.xlim([-2,3])
plt.ylim([-3,6])
plt.plot(x_l,df(x_l),color='skyblue',linewidth=5,alpha=0.6)

values = np.array(x_li)
plt.scatter(x_li,slope_li,color='red',s=100,alpha=0.6)

plt.show()

# 3nd chart

plt.subplot(1,3,3)
plt.title('gradient descent closeup')
plt.xlabel('X',fontsize=16)
plt.grid()
plt.xlim(-0.55,-0.2)
plt.ylim(-0.3,0.8)
plt.plot(x_l,df(x_l),color='skyblue',linewidth=5,alpha=0.8)

values = np.array(x_li)
plt.scatter(x_li,slope_li,color='red',s=300,alpha=0.6)

plt.show()
```
